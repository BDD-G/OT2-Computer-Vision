\[ Q = \sqrt{\frac{2 \cdot C \cdot D}{H}} \]
Following are the fixed values of C and H: C is 50. H is 30. D is the variable whose values should be input to your program in a comma-separated sequence. Example Let us assume the following comma-separated input sequence is given to the program: <code>100,150,180</code>. The output of the program should be: <code>18,22,24</code>.</font>

<div>
    <font color="blue">Hints: If the output received is in decimal form, it should be rounded off to its nearest value (for example, if the output received is <code>26.0</code>, it should be printed as <code>26</code>) In case of input data being supplied to the question, it should be assumed to be a console input.</font>

````python

import math
numbers = input("enter comma separated sequence for D : ")

C=50; H=30
values = (numbers.split(","))

result = []
for i in values:
    D=float(i)
    Q=math.sqrt((2*C*D)/H)
    result.append(str(int(round(Q)))) #str for join, int for just int, round for round off
    
print(",".join(result))
````

    <font color="blue"><b><u>Question 12:</u></b> Write a program which prints all possible permutations and combinations of <code>[1,2,3,4,5,6]</code>.</font>
</div>
<div>
    <font color="blue">Hints: Use <code>itertools.permutations()</code>, <code>itertools.combinations()</code> to get permutations and combinations of list.</font>
</div>


**Permutations:**
Permutations of $n$ elements taken $k$ at a time:

$$P(n, k) = \frac{n!}{(n - k)!}$$

Permutations represent the number of ways to arrange $k$ elements from a set of $n$ distinct elements in a specific order.

**Combinations:**
Combinations of $n$ elements taken $k$ at a time:

$$C(n, k) = \frac{n!}{k! \cdot (n - k)!}$$

Combinations represent the number of ways to select $k$ elements from a set of $n$ distinct elements without considering their order.

source: https://www.coursera.org/learn/combinatorics

````python
from itertools import permutations, combinations

L1 = [1, 2, 3, 4, 5, 6]

print("Permutations = ")
for i in permutations(L1): #permutations, taken FULL length at a time, P(6,6)
    print(i)

#combinations of all lengths
print("\nCombinations = ")
for i in range(1, len(L1) + 1):
    for j in combinations(L1, i):
        print(j)
````

<div>
    <font color="blue">To find the center point (cx, cy) in an image (img), use the moments function as 
shown in the code example:</font>
</div>

```python
M = cv2.moments(img)
if M['m00'] > 0:
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])


````python
image = cv2.imread(r"./data/shapes.bmp")
actual = image.copy() #to display
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Function to isolate shape and find its center
def shape_center(hmin, smin, vmin, hmax, smax, vmax):
    mask = cv2.inRange(hsv_image, np.array([hmin, smin, vmin]), np.array([hmax, smax, vmax]))
    M = cv2.moments(mask)
    if M['m00'] > 0:
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        return (cx, cy), mask
    return None, mask

centers = {} #to store shape and corddi combo dict

# from previous task
shape_values = {"Rectangle": (126, 0, 0, 179, 255, 255),
    "Circular": (110, 100, 100, 130, 255, 255),
    "Triangle": (4, 0, 0, 76, 255, 255),
    "Octagon": (89, 133, 176, 175, 255, 255)}

for shape, hsv_range in shape_values.items():
    center, _ = shape_center(*hsv_range)
    centers[shape] = center #cntr coordi. for each shape

for center in centers.values():
    cv2.circle(image, center, 5, (0, 0, 0), -1) # 5 pixl filled circle in balck

# display time:
display2img_cv2plt(actual, 'actual image',
                   image, 'center of each shape',
                   'Determine the center of each shape using mask info', 
                   image_size=(10, 4), main_title_y=1.1)
````

````python
def Capture_Event(event, x, y, flags, params):
    # If the left mouse button is pressed
    if event == cv2.EVENT_LBUTTONDOWN:
        # Print the coordinate of the 
        # clicked point
        print(f"({x}, {y})")
        
if __name__=="__main__":
    # Read the Image.
    img = cv2.imread(r"./data/persp.jpg", 1)
    # Show the Image
    cv2.imshow('image', img)
    # Set the Mouse Callback function, and call
    # the Capture_Event function.
    cv2.setMouseCallback('image', Capture_Event)
    # Press any key to exit
    cv2.waitKey(0)
    # Destroy all the windows
    cv2.destroyAllWindows()
    
# Source: 
#https://pyseek.com/2022/06/get-coordinates-of-the-clicked-points-in-python/
# https://www.youtube.com/watch?v=rR2KNle0-sw
````

````python
image = cv2.imread(r"./data/persp.jpg")

# coordinates from above step.. corners of atual image !
pts1 = np.float32([[36, 19], [283, 61], [217, 547],[397, 395]])

# width and height of new image should be same as actual image so use Shape!
width, height = image.shape[1], image.shape[0]
pts2 = np.float32([[0, 0],  [width - 1, 0],
    [0, height - 1],[width - 1, height - 1]])

# Compute the perspective transform matrix and apply it
M = cv2.getPerspectiveTransform(pts1, pts2)
transformed_image = cv2.warpPerspective(image, M, (width, height))

#display time
display2img_cv2plt(image, 'actual image',
                   transformed_image, 'transformed image',
                   'perspective transformation by using the four corners', 
                   image_size=(10, 5), main_title_y=0.98)
````
